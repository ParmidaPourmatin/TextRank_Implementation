```python
import hazm
import pprint
from hazm import stopwords_list
```

Open file to read (as in array list) and convert it to string

```python
filename = "mashregh.txt"

with open(filename) as h:
  text_aslist = [line.rstrip('\n') for line in h]

text = ' '.join(text_aslist)
print(text)
```

Now we need to normalize it first.

```python
normalizer = hazm.Normalizer()
normalized_text = normalizer.normalize(text)
print(normalized_text)
```

Creating Sentence Tokens using normalized text

```python
sentence_tokenizer = hazm.SentenceTokenizer()
sentence_tokens = sentence_tokenizer.tokenize(normalized_text)
print(sentence_tokens)
```

The next step is to remove stopwords from each of the sentences.

```python
word_tokenizer = hazm.WordTokenizer()
lexical_tokens = word_tokenizer.tokenize(normalized_text)
print(lexical_tokens)
```

```python
word_tokens_all = []
for w in lexical_tokens:
    if w not in stopwords_list():
        word_tokens_all.append(w)

print(word_tokens_all)
```

```python
word_tokens_bysentence=[[words for words in sentence.split(' ') if words not in stopwords_list()] for sentence in sentence_tokens]
for token in word_tokens_bysentence:
  if token==[]:
    word_tokens_bysentence.remove(token)
mainline = str(word_tokens_bysentence)
print(str(word_tokens_bysentence))
```

```python
with open('AllWordTokens.txt', 'a') as f:
  f.write("[")
#with open('WordTokensBySentence.txt', 'a') as f:
  #f.write("[")


for tok in word_tokens_all:
      with open('AllWordTokens.txt', 'a') as f:
        f.write("'")
        f.write(tok)
        f.write("'")
        f.write(" ")
        f.write(",")
#for tok in word_tokens_bysentence:
      #with open('WordTokensBySentence.txt', 'a') as f:
        #f.write(str(tok))
        #f.write(",")


with open('AllWordTokens.txt', 'a') as f:
  f.write("]")
with open('WordTokensBySentence.txt', 'a') as f:
  f.write(mainline)
```
