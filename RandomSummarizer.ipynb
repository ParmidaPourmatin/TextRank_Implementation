```python
import random
import itertools
```

```python
with open('mashregh.txt') as m:
    hypo = [line.rstrip('\n') for line in m]
# hypothesis file has too many \n so we need ro remove them first
for h in hypo:
  if h=='' :
    hypo.remove(h)

# the first element is the title (the title says what article is about)
#del hypo[0]

print(hypo)
```

```python
newHypo = []

#now we need to split the sentences
for h in hypo:
  temp = h.split(".")
  for t in temp:
    if t=='' :
      temp.remove(t)
  newHypo.append(temp)

# in this step we have list of lists so we need to combine them to remake a single list
newHypo = list(itertools.chain.from_iterable(newHypo))

print(newHypo)
```

```python
# generating our random summary
random_generated_list = random.choices(newHypo, k=4)
random_generated = ' '.join(random_generated_list)
print(random_generated)
```

```python
with open('RandomGenerated.txt', 'a') as f:
          f.write(random_generated)
```
